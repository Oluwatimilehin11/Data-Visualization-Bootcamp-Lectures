```python
# required libraries 
import pandas as pd # data manipulation
import matplotlib.pyplot as plt # main visualization module: matplotlib's pyplot
import seaborn as sns # visualizations' top level. seaborn's style on top of matplotlib
sns.set()
```

```python
# load data. Engine Fuel Types for used cars. Data is collected from 1982 until 2016
df_fuel_engine_types = pd.read_csv("C:\\Users\\phera\\Desktop\\Data Visualization\\Starked Area Charts\\stacked_area_chart_data.csv")
```

```python
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Other"],
             df_fuel_engine_types["Gas"])
plt.show
```
