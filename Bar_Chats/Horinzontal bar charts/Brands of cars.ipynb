```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_used_cars = pd.read_csv('bar_chart_data.csv')
```

```python
df_used_cars
```

```python
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'])
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'])
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'])
plt.yticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'],
       color = 'rgbwymc')
plt.yticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'],
       color = 'midnightblue')
plt.yticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'],
       color = '#4103fc')
plt.yticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'],
       color = '#03fc1c')
plt.yticks(rotation = 45)
plt.title('Cars Listings by Brand')
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.barh(y = df_used_cars['Brand'],
       width = df_used_cars['Cars Listings'],
       color = '#32a852')
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Cars Listings by Brand',fontsize = 16, fontweight = 'bold')
plt.ylabel('Number of Listings',fontsize = 13)
plt.show()
```
