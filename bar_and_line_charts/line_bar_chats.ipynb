```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter #percent formatter will allow us to change the scale of the secondary axis in 
                                                #percentage
    
```

```python
df_kdnuggets = pd.read_csv('bar_line_chart_data.csv')
```

```python
df_kdnuggets
```

```python
fig,ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'k')
```

```python
fig,ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'k')
ax1 = ax.twinx() #twinx allows for two seperate charts to share the x-axis
ax1.plot(df_kdnuggets['Year'],
       df_kdnuggets['Python Users'],
       color = '#b60000',
        marker = 'D')
```

```python
fig,ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'k')
ax.set_ylabel('Number of Participants',
             weight = 'bold')
ax.tick_params(axis = 'y',
              width = 2,
              labelsize = 'large')
ax1 = ax.twinx() #twinx allows for two seperate charts to share the x-axis
ax1.set_ylim(0,1) #set_ylim lets us alter the y-axis range by adding a min and max value
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets['Year'],
       df_kdnuggets['Python Users'],
       color = '#b60000',
        marker = 'D')
ax1.set_ylabel('Python Users',
              color = '#b60000',
             weight = 'bold')
ax1.tick_params(axis = 'y',
               color = '#b60000',
              width = 2,
              labelsize = 'large')
```

```python
fig,ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'k')
ax.set_ylabel('Number of Participants',
             weight = 'bold')
ax.tick_params(axis = 'y',
              width = 2,
              labelsize = 'large')
ax1 = ax.twinx() #twinx allows for two seperate charts to share the x-axis
ax1.set_ylim(0,1) #set_ylim lets us alter the y-axis range by adding a min and max value
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets['Year'],
       df_kdnuggets['Python Users'],
       color = '#b60000',
        marker = 'D')
ax1.set_ylabel('Python Users',
              color = '#b60000',
             weight = 'bold')
ax1.tick_params(axis = 'y',
               color = '#b60000',
              width = 2,
              labelsize = 'large')
ax.set_title('KD Nuggets Survey Python Users (2012 - 2019)', fontsize = '14', weight = 'bold')
plt.show()
```
